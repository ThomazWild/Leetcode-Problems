```python
import numpy as np
mat = [[1,0,1],
      [1,1,0],
      [1,1,0]]

arr = np.array(mat)
n,p = arr.shape

shapes = []
total = 0


#getting all possible shaped of the matrix
for i in range(1, n + 1):
    for l in range(1, p + 1):
        shapes.append((i,l))

print(shapes)

print("-----------------------------------")

for a,b in shapes:
    print("+++++++++")
    print(a,b)
    for rows in range( n-a  +1):
        for cols in range( p-b  +1):
            print(arr[rows:rows+a,cols:cols+b])
            if np.all(arr[rows:rows+a,cols:cols+b] == 1):
                total += 1
            
print("Final answer:", end = " ")
print(total)
#MATRIX LOOPING MODULE
a,b = 3,1
print("-----------------------------------")

for rows in range( n-a  +1):
    for cols in range( p-b  +1):
        print(arr[rows:rows+a,cols:cols+b])
```
